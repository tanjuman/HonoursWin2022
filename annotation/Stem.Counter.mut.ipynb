```python
import glob
```

```python
#Reading in the st files and collecting the stems
def ST_read_in(folderName,countFname):
    countFile = open(countFname,"w")
    ST_files = glob.glob(folderName)

    headerList = ["Name","S","M","I","B","H","K","E","X"]
    headerStr = ""
    for i in headerList:
        headerStr += i + "\t"
    print(headerStr)
    countFile.write(headerStr + "\n")
    
    # paired "Stem"     S - 0
    # Multiloop         M - 1
    # Internal loop     I - 2
    # Bulge 	        B - 3
    # Hairpin loop      H - 4
    # pseudoKnot	    K - 5
    # dangling End	    E - 6
    # eXternal loop     X - 7
    
    for i in ST_files:
        motif_count = [0]*9
        #print(i)
        f = open(i,"r")
        
        count = 0
        
        for j in f:
            nameHead = j.split(" ")
            print(nameHead)
            if count == 0:    
                nameHead = nameHead[4][0:-1]
                print(nameHead)
                motif_count[0] = nameHead
            
            if count > 6:
                firstTerm = nameHead[0]
                #print(firstTerm)
                motifType = firstTerm[0]
                
                if motifType == 's':
                    motif_count[1] += 1
                elif motifType == 'M':
                    motif_count[2] += 1
                elif motifType == 'I':
                    motif_count[3] += 1
                elif motifType == 'B':
                    motif_count[4] += 1
                elif motifType == 'H':
                    motif_count[5] += 1
                elif motifType == 'K':
                    motif_count[6] += 1
                elif motifType == 'E':
                    motif_count[7] += 1
                elif motifType == 'X':
                    motif_count[8] += 1
                    
            
            count += 1 
           
        f.close()
        finStr = ""
        for i in motif_count:
            finStr += str(i) + "\t" 
    
        print(finStr)
        countFile.write(finStr + "\n")
    
ST_read_in("C:\\Users\\tanju\\Documents\\Honours Work\\Outputs\\Hsapi19\\original\\ST\*.st", "Hsapi19.org.count.txt")
```
